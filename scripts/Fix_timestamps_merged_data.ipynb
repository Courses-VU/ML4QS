```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('merged_data.csv')
```

```python
data.head()
```

```python
data["timestamp"] = 0
```

```python
data['time'] = data["time.1"]
data = data.drop("time.1", axis=1)
data.head()
```

```python
timestamp = []
for i in range(0,len(data.index)):
    timestamp.append(pd.to_datetime("6/14/2019 12:06:00") + pd.to_timedelta(data["time"][i]*1000000000, unit = "ns"))
data["timestamp"] = timestamp
```

```python
data.head()
```

```python
data = data.set_index("timestamp")
data.head()
```

```python
data.to_csv("merged_data_fixed.csv")
```
